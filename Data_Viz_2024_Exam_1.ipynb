```python
waterq['month'] = waterq['month'].str.split('_')
```

```python
waterq.head()
```

```python
waterq_2018 = waterq[waterq['month'].apply(lambda x: x[1] == '2018')]
waterq_2019 = waterq[waterq['month'].apply(lambda x: x[1] == '2019')]

mean_chlo_2018 = waterq_2018.groupby('station')['Chlo_a'].mean()
mean_chlo_2019 = waterq_2019.groupby('station')['Chlo_a'].mean()


diff = mean_chlo_2019 - mean_chlo_2018

print(diff)
```
