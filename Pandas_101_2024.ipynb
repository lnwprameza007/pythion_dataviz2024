##Quiz 8
วนลูปนิทานที่มีคำว่า แม่ หรือ พ่อ กี่เรื่อง อะไรบ้าง และคนแต่ง  
มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
คนแต่งคนไหน ชอบแต่นิทานที่มีชื่อเรื่อง พ่อ หรือ แม่มากที่สุด


````
# prompt: ##Quiz 8
# วนลูปนิทานที่มีคำว่า แม่ หรือ พ่อ กี่เรื่อง อะไรบ้าง และคนแต่ง
# มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
# คนแต่งคนไหน ชอบแต่นิทานที่มีชื่อเรื่อง พ่อ หรือ แม่มากที่สุด

# วนลูปนิทานที่มีคำว่า แม่ หรือ พ่อ กี่เรื่อง อะไรบ้าง และคนแต่ง
count_mom_dad = 0
list_mom_dad_stories = []
creator_count = {}
for index, row in folk_df.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count_mom_dad += 1
    list_mom_dad_stories.append((row['title'], row['creator']))  # เก็บชื่อนิทานและคนแต่ง
    if row['creator'] in creator_count:
      creator_count[row['creator']] += 1
    else:
      creator_count[row['creator']] = 1

print(f"มีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ทั้งหมด {count_mom_dad} เรื่อง")
print("รายชื่อนิทานและคนแต่ง:")
for title, creator in list_mom_dad_stories:
  print(f"- {title} (แต่งโดย {creator})")

# มีคำดังกล่าวในนิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน
count_life = 0
count_local = 0
for index, row in folk_df.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']):
    if row['type'] == 'นิทานชีวิต':
      count_life += 1
    elif row['type'] == 'นิทานประจำถิ่น':
      count_local += 1

if count_life > count_local:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานชีวิตมากกว่า")
elif count_local > count_life:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานประจำถิ่นมากกว่า")
else:
  print("มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")

# คนแต่งคนไหน ชอบแต่นิทานที่มีชื่อเรื่อง พ่อ หรือ แม่มากที่สุด
max_count = 0
most_frequent_creator = ""
for creator, count in creator_count.items():
  if count > max_count:
    max_count = count
    most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่อง 'พ่อ' หรือ 'แม่' มากที่สุดคือ {most_frequent_creator} ({max_count} เรื่อง)")
````

##HW6
บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง

(พรีเซ็นคนละ 3 นาที slides)

````
movie_df
````

````
# prompt: ในตารางมีกี่ Genre อะไรบ้าง

# นับจำนวน Genre ที่แตกต่างกัน
num_genres = movie_df['Genre'].nunique()

# แสดงจำนวน Genre
print("จำนวน Genre ทั้งหมด:", num_genres)

# แสดงรายชื่อ Genre ทั้งหมด
all_genres = movie_df['Genre'].unique()
print("\nรายชื่อ Genre ทั้งหมด:")
for genre in all_genres:
  print("- ", genre)
````

````
# prompt: ผู้กำกับที่มี Rating มากที่สุด และคือหนังเรื่องอะไร

# หาผู้กำกับที่มี Rating มากที่สุด
max_rating = movie_df['Rating'].max()
best_rated_directors = movie_df[movie_df['Rating'] == max_rating]['Director'].unique()

# หาหนังของผู้กำกับเหล่านั้นที่มี Rating มากที่สุด
for director in best_rated_directors:
  best_movies_by_director = movie_df[(movie_df['Director'] == director) & (movie_df['Rating'] == max_rating)]['Name']
  print(f"ผู้กำกับ {director} มี Rating สูงสุด ({max_rating}) ในหนังเรื่อง:")
  for movie in best_movies_by_director:
    print(f"- {movie}")
````

````
# prompt: ผู้กำกับคนไหนมี Genre มากที่สุดไม่ซ้ำกัน มี Genre อะไรบ้าง

# สร้าง dictionary เพื่อเก็บจำนวน Genre ของผู้กำกับแต่ละคน
director_genres = {}
for index, row in movie_df.iterrows():
  director = row['Director']
  genre = row['Genre']
  if director not in director_genres:
    director_genres[director] = set()
  director_genres[director].add(genre)

# หาผู้กำกับที่มี Genre มากที่สุดไม่ซ้ำกัน
max_genres = 0
director_with_most_genres = ""
for director, genres in director_genres.items():
  if len(genres) > max_genres:
    max_genres = len(genres)
    director_with_most_genres = director

# แสดงผลลัพธ์
print(f"ผู้กำกับที่มี Genre มากที่สุดไม่ซ้ำกันคือ {director_with_most_genres} ({max_genres} Genre)")
print("Genre ที่กำกับมีดังนี้:")
for genre in director_genres[director_with_most_genres]:
  print(f"- {genre}")
````

````
# prompt: ผู้กำกับคนไหนมีหนังเยอะสุด อะไรบ้าง

# นับจำนวนหนังของผู้กำกับแต่ละคน
director_counts = movie_df['Director'].value_counts()

# หาผู้กำกับที่มีหนังเยอะที่สุด
most_prolific_director = director_counts.index[0]
num_movies = director_counts.iloc[0]

# หาชื่อหนังของผู้กำกับคนนั้น
movies_by_most_prolific_director = movie_df[movie_df['Director'] == most_prolific_director]['Name'].tolist()

# แสดงผลลัพธ์
print(f"ผู้กำกับที่มีหนังเยอะที่สุดคือ {most_prolific_director} ({num_movies} เรื่อง)")
print("รายชื่อหนัง:")
for movie in movies_by_most_prolific_director:
    print(f"- {movie}")
````
