```python
# prompt: barchart เปรียบเทียบปริมาณ sales ของแต่ละ region

import matplotlib.pyplot as plt

# Group data by region and sum sales
region_sales = data.groupby('Region')['Sales'].sum()

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(region_sales.index, region_sales.values)

# Add labels and title
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales by Region')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python
# prompt: สร้าง hist ของ movie โดยใช้ Genre ของ allGenre

import matplotlib.pyplot as plt

# Assuming 'allGenre' is a list of unique genres and 'movie' is your DataFrame

# Create a dictionary to store the count of each genre
genre_counts = {genre: 0 for genre in allGenre}

# Iterate through the 'Genre' column of the DataFrame
for genre_list in movie['Genre'].dropna():
    for genre in genre_list.split(','):
        genre = genre.strip()
        if genre in genre_counts:
            genre_counts[genre] += 1

# Create a bar chart of the genre counts
plt.figure(figsize=(15, 6))
plt.bar(genre_counts.keys(), genre_counts.values())
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Distribution of Movie Genres')
plt.show()
```

```python
# prompt: นับนักแสดง Actor 1 Actor 2 ที่เล่นด้วยกันเยอะที่สุด และplot ออกมาเป็น histrogram โดยแต่ละแท่งจะบอกชื่อนักแสดงสองคน

from collections import defaultdict

actor_pairs = defaultdict(int)

for _, row in movie.iterrows():
    actor1 = row['Actor 1']
    actor2 = row['Actor 2']
    if pd.notna(actor1) and pd.notna(actor2):
        pair = tuple(sorted((actor1, actor2)))
        actor_pairs[pair] += 1

sorted_pairs = dict(sorted(actor_pairs.items(), key=lambda item: item[1], reverse=True))

# Extract top N pairs (e.g., top 10)
top_n = 10
top_pairs = list(sorted_pairs.items())[:top_n]

# Create a list of pair labels and counts for the histogram
pair_labels = [f"{pair[0]} & {pair[1]}" for pair, count in top_pairs]
pair_counts = [count for pair, count in top_pairs]

# Create the histogram
plt.figure(figsize=(10, 6))
plt.bar(pair_labels, pair_counts)
plt.xlabel("Actor Pairs")
plt.ylabel("Number of Movies Together")
plt.title("Top Actor Pairs")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

## Radar Chart

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
data = {'client_id' : ['c1' , 'c2' , 'c3' , 'c4'],
         'cheese' : [2,5,4,3],
         'beer' : [2,1,3,5],
         'chicken' : [5,3,3,4],
        'bread' : [4,3,2,1],
       'coffee' : [2,3,5,3]}
df = pd.DataFrame.from_dict(data)
df
```

```python
categories = list(df.columns[1:])
categories
```

```python
values = df.mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

ax.plot(angles, values, linewidth=1, linestyle='solid' , color = '#ea3033') #เส้นกราฟ
ax.fill(angles, values, 'lightpink', alpha=0.5) #พื้นที่
```

```python
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

# part 1
val_c1 = df.loc[0].drop('client_id').values.flatten().tolist()
val_c1 += val_c1[:1]
ax.plot(angles, val_c1, linewidth=1,
        linestyle='solid', label='Client c1')
ax.fill(angles, val_c1, 'skyblue', alpha=0.4)

# part 2
val_c2=df.loc[1].drop('client_id').values.flatten().tolist()
val_c2 += val_c2[:1]
ax.plot(angles, val_c2, linewidth=1,
        linestyle='solid', label='Client c2')
ax.fill(angles, val_c2, 'lightpink', alpha=0.4)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
