```python
movie_df['Year'] = movie_df['Year'].str.strip('()').astype(int)
```

```python
movie_df['Year']
```

```python
drama_df = movie_df[(movie_df['Genre'].str.contains('Drama')) & (movie_df['Year'] >= 2000) & (movie_df['Year'] <= 2021)]
action_df = movie_df[(movie_df['Genre'].str.contains('Action')) & (movie_df['Year'] >= 2000) & (movie_df['Year'] <= 2021)]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(drama_df['Year'], drama_df['Rating'], label='Drama', alpha=0.5)
plt.scatter(action_df['Year'], action_df['Rating'], label='Action', alpha=0.5)

plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Scatter Plot of Movie Ratings (Drama vs. Action)')
plt.legend()
plt.grid(True)
plt.show()
```
